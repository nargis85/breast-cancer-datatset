# Building Model


```python
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.30,random_state=40)
```

# feature Scaling

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
xtrain=std.fit_transform(xtrain)
xtest=std.fit_transform(xtest)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
model1=lr.fit(xtrain,ytrain)
```

```python
prediction1=model1.predict(xtest)
```

```python
prediction1
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ytest,prediction1)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
plt.savefig('h.png')
```

```python
tp, fn, fp, tn = confusion_matrix(ytest,prediction1).reshape(-1)
```

```python
tp, fn, fp, tn
```

```python
print('Testing Acuracy:',(tp+tn)/(tp+tn+fn+fp))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,prediction1)
```


# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
model2=dtc.fit(xtrain,ytrain)
prediction2=model2.predict(xtest)
cm2=confusion_matrix(ytest,prediction2)
```

```python
cm2
```

```python
tp, fn, fp, tn = confusion_matrix(ytest,prediction2).reshape(-1)
```

```python
tp, fn, fp, tn
```

```python
accuracy_score(ytest,prediction2)
```

# K Nearest Neighbor (K NN)

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
model3=knn.fit(xtrain,ytrain)
prediction3=model3.predict(xtest)
cm3=confusion_matrix(ytest,prediction3)
```

```python
cm3
```

```python
tp, fn, fp, tn = confusion_matrix(ytest,prediction3).reshape(-1)
```

```python
tp, fn, fp, tn
```

```python
#TP :114 cases are correctly identified TN :51 are correctly rejected FN : 1 are incorrectly rejected and FP : 5 are incorrectly identified
```

```python
print(accuracy_score(ytest, prediction3))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, prediction3))
```

```python
#We are getting the best accuracy with knn which is 96.4% , the model is predicting with 96% accuracy on our test data
```
